```python
import os
import pandas as pd

df = pd.DataFrame(columns=['filename','statement'])
for file in os.listdir('P_ZXL_ZFRX_TABLE_MON'):
    sql = open('P_ZXL_ZFRX_TABLE_MON/' + file, 'r', encoding = 'utf-8')
    sqltxt = sql.readlines()
    df = df.append({'filename':file,'statement':sqltxt},ignore_index=True)
    sql.close()
```

```python
import re

# 去除空行空格并大写
df['statement'] = df['statement'].apply(lambda x:' '.join(x))
df['statement'] = df['statement'].apply(lambda x:re.sub('[ ]+',' ',str(x)))
#df['statement'] = df['statement'].apply(lambda x:x.upper())
```

```python
df['statement'] = df['statement'].apply(lambda x:x.split( ))
```

```python
df_word_index1 = pd.DataFrame(columns=['word','index'])
for val in list(enumerate(df.iloc[0,1])):
    df_word_index1 = df_word_index1.append({'word':val[1],'index':val[0]},ignore_index=True)
df_word_index1 = df_word_index1.groupby(['word'],as_index=True)['index'].apply(list)
df_word_index1 = pd.DataFrame(df_word_index1)
df_word_index1
```

```python
df_word_index2 = pd.DataFrame(columns=['word','index'])
for val in list(enumerate(df.iloc[1,1])):
    df_word_index2 = df_word_index2.append({'word':val[1],'index':val[0]},ignore_index=True)
df_word_index2 = df_word_index2.groupby(['word'],as_index=True)['index'].apply(list)
df_word_index2 = pd.DataFrame(df_word_index2)
df_word_index2
```

```python
df_word_index = pd.concat([df_word_index1, df_word_index2], axis=1, join='outer')
df_word_index
```

```python
df_word_index.to_csv('df_word_index.csv',encoding='gb2312')
```
